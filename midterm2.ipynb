## 題目一


<img src="https://drive.google.com/uc?id=17zfTDePcgaRhxk5HAMt93BZPbNpuoYWk"/>

```python
data_list = ["pizza", 100, {"a": 1, "b": 2}, {"pizza": 100, "apple": 50}]

def str_func(data):
    return len(data)

def int_func(data):
    return data ** 2

def dict_func(data):
    keys = list(data.keys())
    values = list(data.values())
    return (keys, values)

def dispatcher(data, str_fn, int_fn, dict_fn):
    if isinstance(data, str):
        return str_fn(data)
    elif isinstance(data, int):
        return int_fn(data)
    elif isinstance(data, dict):
        return dict_fn(data)
    else:
        return "Unsupported data type"

for data in data_list:
    result = dispatcher(data, str_func, int_func, dict_func)
    print("Data:", data)
    print("Result:", result)
    print()
```

## 題目二

<img src="https://drive.google.com/uc?id=1duPC8ayMz2tu88t0YAKLLMbamHfy70Ft"/>

```python
destinations = {
    "Paris": 83000,
    "Rome": 76000,
    "Tokyo": 35000,
    "New York": 100000,
    "Barcelona": 67000,
    "Bali": 15000,
    "Seoul": 32000,
    "Hong Kong": 22000
}

def suggest_trip(destinations, total_budget):

    suitable_destinations = []

    for destination, cost in destinations.items():
        if cost <= total_budget:
            suitable_destinations.append(destination)


    if not suitable_destinations:
        return "沒有在預算內的旅遊目的地"

    suitable_destinations = sorted(suitable_destinations, key=lambda x: destinations[x], reverse=True)

    result = f"預算為{total_budget}可以去的旅遊目的地: {', '.join(suitable_destinations)}"
    return result

total_budget = 80000

trip_suggestion = suggest_trip(destinations, total_budget)
print(trip_suggestion)
```

```python
# step 1 的 dataframe 中每個欄位空值的總數。

from IPython.display import display
null_counts = tb.isnull().sum()

tb['Content_len'] = tb['Content'].apply(lambda x: len(str(x)))

def count_unique_pos(pos_list):
    unique_pos = set()
    for pos in pos_list:
        unique_pos.update(pos)
    return len(unique_pos)

tb['POS_type_num'] = tb['POS'].apply(count_unique_pos)


print(null_counts)

display(tb.head(5))



```

### 3.3

<img src="https://drive.google.com/uc?id=1vJ97Feuqi6ZkvXCd3Z-XL14vYpX9lGKZ"/>

```python
''' 要 print 的內容：
# content_dict這個字典的資料數量（一個整數）

# 在content_dict中找出 盤古開天的神話 這篇文章並列印
'''
content_dict = {}

for index, row in tb.iterrows():
    title = row['Title']
    content = row['Content']
    content_length = row['Content_len']
    num_of_pos = row['POS_type_num']

    content_info = {
        'Content': content,
        'Content_length': content_length,
        'num_of_POS': num_of_pos
    }

    content_dict[title] = content_info

print(f"content_dict 的資料總數: {len(content_dict)}")


target_article = '盤古開天的神話'
if target_article in content_dict:
    print({target_article: content_dict[target_article]})
```

```python
''' 要 print 的內容：
# dialogue_dict 總共有多少篇文章

# 其中有多少篇文章有出現對話

# 生活處處美 這篇文章的對話內容和對話數量
'''
import re

dialogue_dict = {}

for index, row in tb.iterrows():
    title = row['Title']
    content = row['Content']

    dialogues = re.findall(r'「(.*?)」', content)

    num_of_dialogue = len(dialogues)

    dialogue_info = {
        'dialogues': dialogues,
        'num_of_dialogue': num_of_dialogue
    }

    dialogue_dict[title] = dialogue_info

print(f"總共有多少篇文章: {len(dialogue_dict)}")

# 其中有多少篇文章有出現對話
articles_with_dialogues = [article for article, info in dialogue_dict.items() if info['num_of_dialogue'] > 0]
print(f"有出現對話的文章: {len(articles_with_dialogues)}")

# 生活處處美 這篇文章的對話內容和對話數量
target_article = '生活處處美'
if target_article in dialogue_dict:
    print({target_article: dialogue_dict[target_article]})
```
